```python
import spacy
from spacy.displacy import render
from spacy.tokens import DocBin, Span
```

```python
nlp = spacy.load("training/spancat/model-best")
```

```python
docs = list(DocBin().from_disk("corpus/dev-nl.spans.spacy").get_docs(nlp.vocab))
```

```python
doc = docs[2]
```

```python
doc.text
```

```python
doc.spans["sc"] = []
for span in doc.spans["NE-MAIN"]:
    doc.spans["sc"] += [Span(doc, span.start, span.end, label="GOLD_" + span.label_)]
```

```python
doc = nlp(doc)
```

```python
for span in doc.spans["NE-MAIN"]:
    doc.spans["sc"] += [Span(doc, span.start, span.end, label="PRED_" + span.label_)]
```

```python
colors = {"GOLD_Person": "deepskyblue", "GOLD_Organization": "red", "GOLD_Place": "limegreen", "PRED_Person": "lightblue", "PRED_Organization": "pink", "PRED_Place": "lightgreen"}
render(doc, style="span", options={"colors": colors})
```
